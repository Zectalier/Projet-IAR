# Reproducibility Study of Reinforcement Learning Article

## Introduction
This Jupyter Notebook documents a university research project focused on conducting a reproducibility study of a Reinforcement Learning article. The study involves implementing the DroQ algorithm based on REDQ, as the existing library only contains the SAC algorithm. 

Then, we will compare the performance of the DroQ algorithm with the REDQ algorithm, as well as the SAC algorithm. The performance will be measured by the average return of the agent over 2 million steps. To ensure the reproducibility of the results, we ran the experiments 15 times and used the reliable_stats library.

## Differences between REDQ and SAC

The REDQ algorithm is an improvement of the SAC algorithm. The main difference between the two algorithms is that REDQ uses M critic networks instead of 2 in SAC. To compute the actor loss, REDQ uses the average of the M critic networks, while SAC uses the minimum of the 2 critic networks:

$$
\frac{1}{|\mathcal{B}|} \sum_{s \in \mathcal{B}} \left( {\color{red}\frac{1}{N} \sum_{i=1}^{N} Q_{\phi_i}(s, a)} - \alpha \log \pi_\theta(a|s) \right), \quad a \sim \pi_\theta(\cdot|s)
$$

We also use a UTD (Update-to-data) Ratio of 20, which means that we update the critic networks 20 times for each update of the actor network. Which increases the sample efficiency of the algorithm, while a higher number of critic networks reduce the overestimation bias of the Q-values in REDQ.

## Differences between REDQ and DroQ

The DroQ algorithm is an improvement of the REDQ algorithm. The main difference between the two algorithms is that DroQ uses a dropout layer in the critic networks. Thanks to the dropout layer, we can reduce the number of critic networks to 2, while keeping the same performance as REDQ.
